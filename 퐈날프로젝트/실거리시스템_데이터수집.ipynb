```python
import pandas as pd
```

```python
df = pd.read_excel('평택매매2022.xlsx')
df = df[15:]
col = list(df.iloc[0])
df.columns = col
df = df[1:]
```

```python
for i in [2020,2021]:
    df_plus = pd.read_excel(f'평택매매{i}.xlsx')
    df_plus = df_plus[15:]
    col = list(df_plus.iloc[0])
    df_plus.columns = col
    df_plus = df_plus[1:]
    df = df.append(df_plus)
    
df
```
