```python
import pandas as pd
```

```python
df = pd.read_excel('2021가평.xlsx')
df = df[15:]
col = list(df.iloc[0])
df.columns = col
df = df[1:]
df
```

```python
for i in [2018,2019,2020]:
    df_plus = pd.read_excel(f'{i}가평.xlsx',thousands=',')
    df_plus = df_plus[15:]
    col = list(df_plus.iloc[0])
    df_plus.columns = col
    df_plus = df_plus[1:]
    df = df.append(df_plus)
    
df
```

```python
data = df[['계약년월','거래금액(만원)']]
data = data.groupby(['계약년월']).mean()
data
```

```python
data.to_csv('가평군.csv', encoding = 'utf-8-sig')
```
